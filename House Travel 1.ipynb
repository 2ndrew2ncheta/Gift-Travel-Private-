# This notebook is used to scrape ProPublica's database of Privately funded travel in the house of representatives

```python
import pandas as pd
import requests
import time
from random import randrange
from bs4 import BeautifulSoup
```

```python
base_url="https://projects.propublica.org/represent/privately-funded-travel?page="
```

```python
#A list of destinations within Israel
IP_destinations=["israel","tel avi", "jerusalem", "tiberias","tiberas","tiberius", 
                 "isreal", "tibenus",
    "tel-aviv", "israe", "sderot", "jerusalum", "ben gurion", "jersalem", "neve shalom",
     "tiberus","west bank","gaza","ramallah","nablus","jenin", "palestin"]
```

```python
broken_links=[]
```

```python
def scrapepage(url):
    '''
    Scrapes a table from the propublica website into a dataframe and attaches links to any
    pdfs on the site
    '''
    response=requests.get(url)
    if response.status_code>=400:
        broken_links.append(url)
        print(f"failed to scrape page {url}")
    table=pd.read_html(response.text)
    df=table[0]
    soup=BeautifulSoup(response.text, "html.parser")
    table_items=soup.find_all("td")
    #this part creates another column for the pdf links on the page
    links=[]
    for item in table_items:
        link=item.find("a")
        if link!=None:
            if ".pdf" in link.get("href"):
                links.append(link)
    df["hrefs"]=[link.get("href") for link in links]
    time.sleep(1)
    return df
```

### Now we start by scraping the first page

```python
alldata=scrapepage(base_url+"1")
alldata["Page #"]=1
alldata
```

### Now we loop through the next 602 pages

```python
for i in range(2,153):
    print(f"now scraping page {i} of 152")
    page=scrapepage(f"{base_url}{i}")  
    page["Page #"]=i
    alldata=pd.concat([alldata,page])
print("Now done scraping")
```
