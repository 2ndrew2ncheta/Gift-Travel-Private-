```python
Israel_destinations=["israel","tel avi","tel aviv", "jerusalem", "tiberias","tiberas","tiberius", 
                 "isreal", "tibenus","tlv","tel",
    "tel-aviv", "israe", "sderot", "jerusalum", "ben gurion", "jersalem", "neve shalom",
     "tiberus"]#some filers just list the city, so it's important to identify every possible 
#way of saying the same thing. 
```

```python
Palestine_destinations=["west bank","gaza","ramallah","nablus","jenin", "palestin"]
```

```python
def findcollisions(mystring, mylist):
    '''
    searches a string 'mystring' for elements listed in [mylist]. Returns all 
    elements separated by a space
    '''
    stringlist=mystring.split(',') #changes mystring to list form, splitting at commas
    text=""
    for item in mylist:
        if item in mystring:
            text=text+" "+item
    return text.strip()
```

```python
def identify_country(string):
    '''identifies countries based on the established lists of countries and states.
    assumes that no string has locations in both the US and a foreign country. 
    If the function cannot identify the destination country based on those lists, it returns 
    the original string
    '''
    Israel=False
    Palestine=False

    string=string.replace("Britain","United Kingdom").replace("UK","United Kingdom")\
    .replace("England", "United Kingdom").replace("Scotland","United Kingdom")\
    .replace("United Kingdon","United Kingdom").replace("United Kington","United Kingdom")
    string=string.replace("Ottawa","Canada").replace("Calgary","Canada")
    string=string.replace("Columbia","Colombia") #This section removes some common misspellings.
    string=string.replace("United Arab Emirates","UAE")
    if any(x in string.lower() for x in Palestine_destinations):
        Palestine=True
    else:
        pass
    if any(x in string.lower() for x in Israel_destinations):
        Israel=True
    else:
        pass
    print(Israel)
    print(Palestine)
        
    if Israel==True and Palestine==True:
        return "Israel/Palestine"
    elif Israel==True:
        return "Israel"
    elif Palestine==True:
        return "Palestine"
    elif any(x in string for x in list_of_countries):
        return findcollisions(string, list_of_countries)
    elif any(x in string for x in list_of_states):
        return "USA"
    else:
        return string
```

```python
alldata[["Departure","Return"]]=alldata.Dates.str.split('-',expand=True)
alldata["Year"]=pd.DatetimeIndex(alldata["Departure"]).year
```

```python
alldata["Country"]=alldata["Destination"].apply(lambda x: identify_country(x))
```
